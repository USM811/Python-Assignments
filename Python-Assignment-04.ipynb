## Q.1- Make a calculator using Python with addition , subtraction , multiplication ,division and power.

```python
check_point = True
n=1
while(check_point):
    a = float(input("Enter the first number: "))
    b = float(input("Enter the second number: "))
    
    if n==1:
        print("Press 'A' to calculate the addition.")
        print("Press 'S' to calculate the subtraction.")
        print("Press 'M' to calculate the multiplication.")
        print("Press 'D' to calculate the division - 1st number / 2nd number.")
        print("Press 'P' to calculate the (a^b) 1st_number-rais-to-power-2nd_number.")
    operation = input()
    
    if operation=='a' or operation=='A':
        _sum = a + b
        print( str(a) + " + " + str(b) + " = " + str(_sum))
        n+=1
    elif operation=='s' or operation=='S':
        _sub = a - b
        print( str(a) + " - " + str(b) + " = " + str(_sub))
        n+=1
    elif operation=='m' or operation=='M':
        _mul = a * b
        print( str(a) + " * " + str(b) + " = " + str(_mul))
        n+=1
    elif operation=='d' or operation=='D':
        _div = a / b
        print( str(a) + " / " + str(b) + " = " + str(round(_div,2)))
        n+=1
    elif operation=='p' or operation=='P':
        _pow = a ** b
        print( str(a) + " ^ " + str(b) + " = " + str(_pow))
        n+=1
    else:
        print("You Entered Wrong Input for the Given Operations.")
    
    again = input("Press 'Y' to use the calculator again, or Press any other key to exit.")
    if again=='y' or again=='Y':
        check_point = True
    else:
        check_point = False
```

## Q.2- Write a program to check if there is any numeric value in list using for loop.

```python
my_list = ['Arham', '8', 'Salman', '11', 'Marya', '12']

number_of_numeric_values = 0
numeric_list = []

for item in my_list:
    try:
        numeric_list.append(float(item))
        number_of_numeric_values +=1
    except ValueError:
        continue

print("There are " + str(number_of_numeric_values) + " numerical values in the list.")
print("List of Numerical Values are " , numeric_list)
```

## Q.3- Write a Python script to add a key to a dictionary.

```python
my_dict = {}

n = int(input("How many Items do you want to Add in the Dictionary? "))
check_point = 0
while check_point < n:
    key = input("Enter the 'Key' of New Item: ")
    value = input("Enter the 'Value' of the corresponding 'Key': ")
    my_dict.update({key: value})
    check_point +=1
print("\nYou added " + str(check_point) + " items in the dictionary.")
print("\nThe updated dictionary is =>" , my_dict)
```

## Q.4- Write a Python program to sum all the numeric items in a dictionary.

```python
rate_list = {'Cherry': 'Too Expensive','Apple': '250', 'Banana': '100', 'Orange': '60', 'Mango': 150, 'Melon': 'Not-Available'}

list_of_rates = []

for i in rate_list.values():
    try:
        list_of_rates.append(float(i))
    except ValueError:
        continue
print("From the Dictionary, List of Numerical Values is =>", list_of_rates)
print("\nSum of Numerical Values from the Dictionary is = " + str(sum(list_of_rates)))
```

## Q.5- Write a program to identify duplicate values from list.

```python
random_list = ['usman', 30, 'arshman', 22, 'yasra', 28, 'usman', 28, 'mansoor', 30]
repeated_elements = []
n = 0
for i in range(len(random_list)):
    for j in range(i+1, len(random_list)):
        if random_list[i]== random_list[j] and i not in repeated_elements:
            repeated_elements.append(random_list[i])
            n +=1
        else:
            pass
print("There are " + str(n) + " elements repeating in the list.")
print("\nList of repeated elements is =>", repeated_elements)
```
